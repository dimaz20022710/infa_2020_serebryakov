```python
#2
t.reset()

t.penup()

Instructions = ['101111100', '000001110', '101101001', '110000011', '010101100',
                '111100100', '011010110', '100010010', '111111100', '110101001']

Tempstring = 'a'
Posy = [0, -20, -40, 0, -20, 0, -20, 0, -20]
Posx = [0, 0, 0, 0, 0, 20, 20, 20, 20]

def index(number):
    for i in range (len(number)):
        Templist = Instructions[int(number[i])]
        for j in range(3):
            t.goto(i*30+Posx[j], Posy[j])
            t.pendown()
            t.forward(20*int(Templist[j]))
            t.penup()
        t.right(90)
        for j in range(3, 7):
            t.goto(i*30+Posx[j], Posy[j])
            t.pendown()
            t.forward(20*int(Templist[j]))
            t.penup()
        t.right(45)
        for j in range(7, 9):
            t.goto(i*30+Posx[j], Posy[j])
            t.pendown()
            t.forward(28*int(Templist[j]))
            t.penup()
        t.left(135)


index('141700')

```

```python
#4
t.reset()

def physics(x=50, y=100, Vy=10, Vx=-10, Eloss=1):
    t.penup()
    t.goto(x, y)
    t.pendown()
    dt = 0.1
    ay = -10
    for i in range(250):
        if y <= 0:
            Vy = -Vy*Eloss
        x += Vx*dt
        y += Vy*dt + ay*dt**2/2
        Vy += ay*dt
        t.goto(x, y)

physics()
```
